# Brazillian Ecommerce, Olist Dataset

## Dataset Context
This is a Brazilian ecommerce public dataset of orders made at Olist Store.  The dataset has information of 100k orders from 2016 to 2018 made at multiple marketplaces in Brazil. Its features allow viewing an order from multiple dimensions: from order status, price, payment and freight performance to customer location, product attributes and finally reviews written by customers. They also released a geolocation dataset that relates Brazilian zip codes to lat/lng coordinates.

Olist connects small businesses from all over Brazil to channels without hassle and with a single contract. Those merchants are able to sell their products through the Olist Store and ship them directly to the customers using Olist logistics partners. See more on our website: www.olist.com

After a customer purchases the product from Olist Store a seller gets notified to fulfill that order. Once the customer receives the product, or the estimated delivery date is due, the customer gets a satisfaction survey by email where he can give a note for the purchase experience and write down some comments.

# 1. Business Problem Understanding

**1.1 Background**

Olist, a leading e-commerce platform in Brazil, connects small and medium-sized businesses (SMBs) to a vast network of customers across the country. Despite its rapid growth, Olist faced significant challenges in understanding the diverse needs of its customers, efficiently allocating resources, and delivering personalized marketing experiences. To address these issues, Olist embarked on a transformative journey to leverage customer segmentation, aiming to gain deeper insights into customer behavior, improve targeting, and optimize resource allocation.

One of the key methods we adopted for customer segmentation was RFM (Recency, Frequency, Monetary) analysis. This approach allowed Olist to analyze customer behaviors at a granular level, identifying high-value customers who frequently engage with the platform, those who have made recent purchases, and those who have spent significant amounts. By clustering customers into segments based on their RFM scores, Olist could make more strategic decisions regarding product placement, inventory management, and customer engagement. This deeper understanding of customer needs enabled the platform to tailor its offerings and improve overall customer satisfaction.

Moreover, RFM segmentation empowered Olist to enhance its targeting capabilities and allocate resources more efficiently. By identifying distinct customer segments, the platform could design and deliver tailored promotions, significantly increasing the return on investment for marketing campaigns. This targeted approach not only optimized resource allocation but also ensured that marketing efforts were directed toward the most responsive and valuable customer groups.

Finally, RFM analysis enabled Olist to personalize its marketing efforts, creating customized offers and experiences for specific customer segments. This personalization increased the likelihood of conversion and fostered long-term customer loyalty, as customers felt more valued and understood. Through the strategic implementation of RFM segmentation, Olist was able to drive long-term revenue growth while strengthening its relationships with both customers and SMBs on its platform. 

**1.2 Problem Statement**

Understanding customer behavior is crucial for optimizing marketing efforts. By segmenting customers based on key behaviors, such as frequency, recency, and monetary value of their purchases (RFM), Olist can target specific customer groups with personalized offers, improving customer satisfaction and retention. **The challenge is to identify the best segmentation strategy to enhance the customer experience and maximize conversion rates.** 

**Business Problem:**

This lack of customer segmentation results in missed opportunities to address unique preferences and pain points, inefficient resource allocation, and generic marketing campaigns that fail to engage distinct customer groups.

**Why solving this problem is crucial?**

The lack of customer segmentation faced by Olist is important to address due to several critical reasons that directly impact business growth and customer satisfaction:
1. **Enhancing Customer Satisfaction and Loyalty**: Without proper segmentation, Olist cannot understand the unique preferences and needs of different customer groups. This results in a less personalized customer experience, which can reduce satisfaction and loyalty. By implementing segmentation, Olist can deliver more relevant and engaging experiences, thereby improving customer retention.
2. **Optimizing Resource Allocation**: Without a deep understanding of customer behavior, marketing and operational resources may be allocated inefficiently. Segmentation allows Olist to focus on the most valuable or high-potential customer groups, increasing the ROI (Return on Investment) of marketing campaigns and resource utilization.
3. **Improving Marketing Campaign Effectiveness**: Generic marketing campaigns often fail to capture customer attention because they are not tailored to their specific needs or interests. Through segmentation, Olist can design more personalized and targeted campaigns, which in turn increases conversion rates and customer engagement.

By addressing this issue, Olist can not only enhance customer satisfaction and operational efficiency but also drive sustainable business growth and strengthen its competitive edge in the Brazilian e-commerce market.

**1.3 Goals**

The goal of this project is to segment customers into meaningful groups using the RFM model based on their transactional behavior. By clustering customers into distinct segments based on Recency, Frequency, and Monetary values, we aim to:

* Segment customers into distinct groups based on their RFM scores (e.g., recent, frequent, high-spending customers).
* Identify key customer segments that show different engagement patterns and behaviors, such as loyal customers, at-risk customers, and potential high-value customers.
* Provide actionable insights for marketing teams to design personalized campaigns, loyalty programs, and promotional strategies that improve customer satisfaction and increase retention.

**1.4 Stakeholders**

* Marketing Teams: They aim to use customer segmentation to design targeted campaigns, personalized offers, and retention strategies, improving conversion rates and customer loyalty.
* Sales Teams: Sales teams can leverage customer segmentation to identify potential high-value customers, offer personalized sales assistance, and optimize sales efforts.
* Business Owners and Managers: Interested in improving customer engagement and long-term revenue growth by enhancing marketing strategies and resource allocation.
* Data Science Team: Responsible for developing and deploying the RFM segmentation model and analyzing customer behavior to derive actionable insights.

**1.5 Metric Evaluation**

The evaluation metrics for this project include:
1. Silhouette Score: Measures how well the customer segments are defined based on their RFM scores. A higher silhouette score indicates better-defined and actionable segments, ensuring that customers within the same cluster are similar to each other while being distinct from those in other clusters.
2. Elbow Method: This technique is used to determine the optimal number of clusters in a dataset by plotting the within-cluster sum of squares (WCSS) against the number of clusters. The "elbow point" in the graph, where the rate of decrease in WCSS sharply slows down, indicates the ideal number of clusters. This helps balance the trade-off between the complexity of the model (more clusters) and the effectiveness of segmentation (clear, distinct groups).

These metrics collectively ensure that the segmentation strategy is both meaningful and actionable, enabling Olist to create well-defined customer groups for targeted marketing and resource optimization.

**1.6 Business Metrics**

The evaluation used to assess the profit or savings achieved after implementing the model is conducted using the following two methods:
* Customer Retention Rate by Segment: The percentage of customers in each segment that return for repeat purchases, indicating the success of retention efforts.
* Revenue per Segment: The average revenue generated from each customer segment, helping to identify high-value segments for targeted marketing.
* Marketing Campaign Conversion Rate: The conversion rate of marketing campaigns targeting specific segments, measuring how well the campaigns perform for different customer groups.
* Customer Lifetime Value (CLV): Estimated revenue that each segment generates over time, helping businesses prioritize high-value segments.

**1.7 Analytical Approach**

We will apply RFM analysis to segment customers based on three key metrics:

* Recency: How recently did the customer make a purchase? This can indicate customer engagement and loyalty.
* Frequency: How often does the customer make purchases? Frequent buyers are usually more loyal and responsive to offers.
* Monetary: How much does the customer spend? High-spending customers are crucial for maximizing revenue.

After calculating the RFM scores for each customer, we will perform clustering (e.g., K-Means or DBSCAN) to group customers with similar RFM profiles. These customer segments will help marketing and sales teams target the right customers with personalized offers, promotions, and campaigns.

## 4.2 Handling Outliers

For features with extreme values, you can either remove the outliers or cap them. From the distribution of the column, we can see that the distribution of the data columns shows a normal distribution. Hence why, we use ±3 sigma method.

```python
# Remove outliers using ±3 sigma
def remove_outliers(df, column):
    mean = df[column].mean()
    sigma = df[column].std()
    lower_bound = mean - 3 * sigma
    upper_bound = mean + 3 * sigma
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers(df, 'price')
df = remove_outliers(df, 'freight_value')
df = remove_outliers(df, 'payment_value')
```

###  Champions

Description: Customers who are loyal, engaged, and spend frequently and significantly. They have recently made purchases and are likely to buy again.

RFM Criteria:
* Recency (R): 5 (bought recently)
* Frequency (F): 4–5 (buy often)
* Monetary (M): 4–5 (spend a lot)

### Loyal Customers

Description: Customers who purchase regularly and spend reasonably well, but might not engage as frequently as champions.

RFM Criteria:
* Recency (R): 5 (bought recently)
* Frequency (F): 3–4 (buy often)
* Monetary (M): 3–5 (spend reasonably)

### Potential Loyalists

Description: Newer customers who have spent well recently, showing good potential for becoming loyal, but need more engagement.

RFM Criteria:
* Recency (R): 4-5 (bought recently)
* Frequency (F): 2–3 (buy occasionally)
* Monetary (M): 2–4 (spend a moderate-high amount)

### At Risk Customers

Description: Customers who bought in the past, but their engagement is declining. They need to be re-engaged before they churn completely.

RFM Criteria:
* Recency (R): 4 (have not bought recently)
* Frequency (F): 2-4 (buy infrequently)
* Monetary (M): 1–5 (spent previously but less frequently now)

### New Customers

Description: Customers who have recently made their first purchase. They are characterized by having made only one purchase, but this purchase was recent. 

RFM Criteria:

* Recency (R): 5 (bought recently)
* Frequency (F): 0-1 (purchased only once)
* Monetary (M): 1–5 (moderate or low spending)

### Lost Customers

Description: Customers who have not bought in a long time and have low engagement. They have the lowest likelihood of returning unless targeted with a special campaign.

RFM Criteria:
* Recency (R): <4 (haven’t bought in more than 180 days)
* Frequency (F): 0–1 (low interaction)
* Monetary (M): 0–2 (minimal spend or none)

```python
# Ensure that all RFM scores are integers
rfm_df['r_score'] = rfm_df['r_score'].astype(int)
rfm_df['f_score'] = rfm_df['f_score'].astype(int)
rfm_df['m_score'] = rfm_df['m_score'].astype(int)

# Calculate the RFM score (just to make sure it's calculated correctly)
rfm_df['rfm_score'] = 100 * rfm_df['r_score'] + 10 * rfm_df['f_score'] + rfm_df['m_score']

def customer_segmentation(r_score, f_score, m_score):
    # Cluster 1: Champions
    if r_score == 5 and f_score == 5 and m_score == 5:
        return 'Champions'  # High engagement, frequent purchases, and high spend
    
    # Cluster 2: Loyal Customers
    elif r_score >= 4 and f_score >= 4 and m_score >= 3:
        return 'Loyal Customers'  # Regular purchases, good spending
    
    # Cluster 3: Potential Loyalists
    # Recently acquired customers with high recency and moderate frequency (f_score > 1)
    elif r_score >= 4 and f_score in [2, 3, 4] and m_score in [2, 3, 4, 5]:
        return 'Potential Loyalists'  # Recently acquired, moderate frequency and spending
    
    # Cluster 4: At Risk Customers
    # Low recency but some frequency, and moderate past spend
    elif r_score == 4 and f_score in [2, 3, 4] and m_score >= 1:
        return 'At Risk Customers'  # Low recency, but moderate frequency and spending
    
    # Cluster 5: Lost Customers
    # Very low recency and frequency, and minimal spend (likely abandoned)
    elif r_score < 4 and f_score <= 5 and m_score <= 5:
        return 'Lost Customers'  # Low recency, low frequency, low or no spend
    
    # Cluster 6: New Customers
    # Recently bought (Recency = 5), purchased only once (Frequency = 1), low or moderate spend
    elif r_score >= 4 and f_score == 1 and m_score <= 5:
        return 'New Customers'  # Recently bought, moderate/low spend
    
    # Default for any other case (fallback to 'Potential Loyalists')
    return 'Potential Loyalists'  # This fallback assigns any unclassified customer to Potential Loyalists

# Apply the segmentation logic
rfm_df['customer_segmentation'] = rfm_df.apply(
    lambda x: customer_segmentation(x['r_score'], x['f_score'], x['m_score']),
    axis=1
)

# Calculate the percentage of each cluster
cluster_percentage = rfm_df['customer_segmentation'].value_counts(normalize=True) * 100

# Display the percentage breakdown of each cluster
print("Percentage breakdown of each cluster:")
print(cluster_percentage)

# Display the updated DataFrame (first 20 rows for simplicity)
print(rfm_df[['rfm_score', 'customer_segmentation']].head(20))
```

Using elbow method to determine the best cluster number.

```python
# Calculate inertia for different numbers of clusters
inertia = []
for i in range(1, 11):  # Testing from 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(rfm_robust)
    inertia.append(kmeans.inertia_)

# Plot inertia to find the "elbow"
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

Key Observations from the Elbow Plot:

* Inertia Decreases Rapidly: Initially, as the number of clusters increases, inertia drops sharply, which indicates that the clusters are becoming more compact and well-defined.
* Elbow Point: The plot shows a noticeable "elbow" around 3 clusters. This is where the inertia reduction begins to slow down significantly. The elbow indicates that adding more clusters doesn't substantially improve the clustering (in terms of reducing inertia). This is the point where adding additional clusters becomes less beneficial, so 3 clusters would be the optimal number of clusters based on the Elbow Method.

Interpretation:

* Optimal Number of Clusters: The plot suggests that 3 clusters is likely the best choice for segmenting your customers. Beyond this point, the improvement in inertia is minimal, meaning that increasing the number of clusters will not significantly improve the quality of the clustering.
* Cluster Characteristics: After deciding on 3 clusters, you can perform the K-Means clustering with 3 clusters, and then analyze the segments to understand customer behaviors and tailor your marketing and sales strategies accordingly.

## Comparison of Model's Performance

```python
# Function to evaluate clustering models
def evaluate_model(model, data, labels):
    if len(np.unique(labels)) > 1:  # Ensure there are at least 2 clusters
        silhouette = silhouette_score(data, labels)
        calinski = calinski_harabasz_score(data, labels)
        davies = davies_bouldin_score(data, labels)
    else:
        silhouette, calinski, davies = np.nan, np.nan, np.nan
    return silhouette, calinski, davies

# Initialize a DataFrame to store results
results = pd.DataFrame(columns=['Model', 'Silhouette Score', 'Calinski-Harabasz Score', 'Davies-Bouldin Score'])

# List of models to evaluate (excluding MeanShift)
models = {
    'KMeans': KMeans(n_clusters=4, random_state=42),
    'GMM': GaussianMixture(n_components=3, random_state=42),
    'HDBSCAN': hdbscan.HDBSCAN(min_samples=5, min_cluster_size=50)  
}

# Evaluate each model
for model_name, model in models.items():
    # Fit the model and predict cluster labels
    if model_name == 'HDBSCAN':  # Special handling for HDBSCAN, which can produce noise (-1)
        labels = model.fit_predict(rfm_robust)
    else:
        labels = model.fit_predict(rfm_robust)
        
    # Evaluate the model
    silhouette, calinski, davies = evaluate_model(model, rfm_robust, labels)
    
    # Append results to the DataFrame
    results = pd.concat([results, pd.DataFrame({
        'Model': [model_name],
        'Silhouette Score': [silhouette],
        'Calinski-Harabasz Score': [calinski],
        'Davies-Bouldin Score': [davies]
    })], ignore_index=True)

# Display results
print("Model Comparison:")
print(results)

# Plot the results
fig, axes = plt.subplots(3, 1, figsize=(10, 12))
metrics = ['Silhouette Score', 'Calinski-Harabasz Score', 'Davies-Bouldin Score']
for i, metric in enumerate(metrics):
    axes[i].bar(results['Model'], results[metric], color='skyblue')
    axes[i].set_title(metric)
    axes[i].set_ylabel('Score')
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
